```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
from matplotlib.pylab import rcParams
rcParams['figure.figsize'] = 10, 6
```

```python
dataset1 = pd.read_csv('Bitcoin Energy Consumption.csv')
#Convert into datetime format
dataset1['Date'] = pd.to_datetime(dataset1['Date'], infer_datetime_format=True)
indexedDataset1 = dataset1.set_index(['Date'])
```

```python
dataset2 = pd.read_csv('Bitcoin USD (BTC-USD).csv')
#Convert into datetime format
dataset2['Date'] = pd.to_datetime(dataset2['Date'], infer_datetime_format=True)
indexedDataset2 = dataset2.set_index(['Date'])
```

```python
#Determining rolling statistics
#Rolling mean
rolmean = indexedDataset1['Estimated TWh per Year'].rolling(window=365).mean()

#Rolling standard deviation
rolstd = indexedDataset1['Estimated TWh per Year'].rolling(window=365).std()
print(rolmean, rolstd)

#Plot rolling statistics:
orig = plt.plot(indexedDataset1['Estimated TWh per Year'], color='blue',label='Original')
mean = plt.plot(rolmean, color='red', label='Rolling Mean')
std = plt.plot(rolstd, color='black', label='Rolling Std')
plt.legend(loc='best')
plt.title('Rolling Mean & Standard Deviation')
plt.show(block=False)
```
